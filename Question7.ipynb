```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[15, 10])

# Marriages and divorces per capita in the U.S. between 1900, 1950, and 2000
marriages = [709000, 1667000, 2315000]
divorces = [56000, 385000, 944000]

X = np.arange(len(marriages))

plt.bar(X, marriages, color = 'green', width = 0.25)
plt.bar(X + 0.25, divorces, color = 'orange', width = 0.25)

plt.legend(['Marriages', 'Divorces'])

plt.xticks([i + 0.25 for i in range(3)], ['1900', '1950', '2000'])

plt.title("Column Chart for Marriages and Divorces per-capita in the U.S. Between 1900, 1950, and 2000")

plt.xlabel('Marriage-Divorce-per-Capita between 1900, 1950, and 2000')
plt.ylabel('Total')

plt.show()
```
